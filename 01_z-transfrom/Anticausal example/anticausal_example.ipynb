```python
import cv2 as cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Load input image and kernel

```python
image = cv2.imread("whereswally.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = np.array(image, dtype=float) - (image.max() - image.min())/2
kernel = cv2.imread("kernel.png")
kernel = cv2.cvtColor(kernel, cv2.COLOR_BGR2GRAY)
kernel = np.array(kernel, dtype=float)
kernel = kernel - (kernel.max() - kernel.min())/2
delta = cv2.imread("delta.png")
delta = cv2.cvtColor(delta, cv2.COLOR_BGR2GRAY)
delta = np.array(delta, dtype=float)
```

```python
_, ax = plt.subplots(1,1,figsize=(20,20))
ax.imshow(image, cmap='gray')
```

# Apply filter to whereswally

```python
x = image

y = cv2.filter2D(src=x, ddepth=-1, kernel=kernel)
```

```python
row_col_max = np.unravel_index(np.argmax(y), y.shape)
print(row_col_max, ' of ', y.shape)
```

```python
_, ax = plt.subplots(1,1,figsize=(20,20))
ax.imshow(y, cmap='gray')
ax.scatter(row_col_max[1], row_col_max[0], s=1000, c='white', alpha=0.2, edgecolors='blue')
```

# Apply filter to impulse

```python
x_ = delta
_, ax = plt.subplots(1,1,figsize=(20,20))

ax.imshow(x_, cmap='gray')
```

```python
y_ = cv2.filter2D(src=x_, ddepth=-1, kernel=kernel)
```

```python
_, ax = plt.subplots(1,1,figsize=(20,20))
ax.imshow(y_, cmap='gray')
```
